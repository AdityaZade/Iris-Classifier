```python
from sklearn import datasets
from sklearn import tree
```

```python
#Step 1: Loading dataset
iris=datasets.load_iris()


X=iris.data #features
y=iris.target#labels
```

```python
#Step 2: splitting the dataset into train and test data
from sklearn.model_selection import train_test_split

train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.5)
```

```python
#Step 3: Selecting the model and training it 
my_classifier=tree.DecisionTreeClassifier(random_state=17122)

my_classifier.fit(train_X,train_y)
```

```python
#step 4 :Making prediction
predictions=my_classifier.predict(test_X)
```

```python
#Step 5: Checking accuracy of model

from sklearn.metrics import accuracy_score
accuracy=accuracy_score(test_y,predictions)
print(accuracy*100)
```

```python
#Step 6: visiualising the decision tree
import graphviz
# DOT data
dot_data = tree.export_graphviz(my_classifier, out_file=None, 
                                feature_names=iris.feature_names,  
                                class_names=iris.target_names,
                                filled=True)

# Draw graph
graph = graphviz.Source(dot_data, format="png") 
graph
```
